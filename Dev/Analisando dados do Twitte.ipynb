```python
#importando o modulo pandas para trablahar com datasets em python
import pandas as pd
```

```python
#criando um datafrane a partir do dataset
df = pd.DataFrame(dataset)
```

```python
#imprinindo o datafram
df
```

```python
#importanto o modulo scikit learn 
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#usando o metodo Vercotizer para criar uma matriz de documentos
cv = CountVectorizer()
count_matrix = cv.fit_transform(df.text)
```

```python
#contanto o numero de ocorrencias das principais palavras no dataset
word_count = pd.DataFrame(cv.get_feature_names(), columns=['word'])
word_count['count'] = count_matrix.sum(axis=0).tolist()[0]
word_count = word_count.sort_values('count', ascending=False).reset_index(drop=True)
word_count[:50]
```
