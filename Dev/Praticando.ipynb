```python
import matplotlib.pyplot as plt
```

```python
x = [1, 3, 5, 7, 9]
y = [2, 3, 7, 1, 0]
```

```python
titulo = "Scatterplot: grafico de dispersao"
eixox = "Eixo X"
eixoy = "Eixo Y"
```

```python
#legendas
plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)
```

```python
plt.plot(x, y)
plt.scatter(x, y, label="Meus pontos", color="red", marker="^")
plt.legend()
plt.show()
```

```python
#open archives
dados = open("populacao-brasileira.csv").readlines()
```

```python
#making loop for 
x = []
y = []
for i in range(len(dados)):
    if i != 0:
        linha = dados[i].split(";")
        x.append(int(linha[0]))
        y.append(int(linha[1]))
```

```python
#show the graphs
plt.plot(x, y)
plt.bar(x, y, color="#FF00FF")
plt.show()
```

```python
#manipuling boxplot
import random

vetor = []

for i in range(100):
    num = random.randint(0, 50)
    vetor.append(num)

plt.boxplot(vetor)
plt.show()
```
